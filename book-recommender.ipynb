POPULARITY BASED RECOMMENDED SYSTEM

```python
# top 50 books -> highest rating,minimum 250 peoples have voted
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
# popular_df = popular_df.drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_rating']]
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

Collaborative Filtering Based Recommender System

```python
# ek user neh ek book ko kya rating di hai
# user->200 books ko rating kiya hai toh consider
# book ->50 user neh rating kiya toh consider
```

```python
ratings_with_name.head()
```

```python
ratings_with_name.groupby('User-ID')['Book-Rating'].count()
```

```python
ratings_with_name.groupby('User-ID')['Book-Rating'].count()>200
```

```python
filtered_users = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
filtered_users = filtered_users[filtered_users]
```

```python
users_req=filtered_users.index
users_req
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(users_req)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title')['Book-Rating'].count()>=50
famous_book=y[y].index
famous_book
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```
